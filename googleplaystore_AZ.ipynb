```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
csv = pd.read_csv("googleplaystore.csv")
csv = csv.dropna(how = "any")
```

```python
genres_counts = csv.drop_duplicates("Genres")
```

```python
genres_counts["Genres"].count()
```

```python
free_app = csv.loc[csv['Type'] == 'Free']
paid_app = csv.loc[csv['Type'] == 'Paid']
```

```python
print(f'{free_app.count()}')
print(f'{paid_app.count()}')
```

```python
total_free_app = free_app["App"].count()
total_paid_app = paid_app["App"].count()
print(total_free_app+total_paid_app)
```

```python
labels = ["Free App", "Paid App"]
sizes = [total_free_app, total_paid_app]
colors = ["green","orange"]
explode = (0.1,0)

plt.pie(sizes, explode = explode, labels= labels, colors = colors, autopct="%1.1f%%", startangle = 120)
plt.axis("equal")
```
