```python
import pandas as pd
import json
import matplotlib.pyplot as plt
import numpy as np
import requests
import csv
import play_scraper as gp
from pprint import pprint
```

```python
# getting the basic info of the top 50 new free games in the action genre
# note: we can change the subgenre/category/paid vs free/etc and define how many results we want.
action_games = gp.collection(collection='NEW_FREE', category='GAME_ACTION', results=50)
```

```python
# now we can use a loop that will pull the app id's we want from the games we retrieved
# from the search above and store them in a list
i = 0
action_game_list = []
for game in action_games:
    game_info = action_games[i]["app_id"]
    action_game_list.append(game_info)
    i += 1
```

```python
# from here we can loop through the list of games above and isolate the detailed information
# that we need for each game in the list.
i = 0
action_game_titles = []
action_game_scores = []
action_game_reviews = []
action_game_prices = []
action_game_installs = []
action_game_developers = []
action_game_urls = []

for game in action_game_list:
    game_details = gp.details(action_game_list[i])
    action_game_titles.append(game_details['title'])
    action_game_scores.append(game_details['score'])
    action_game_reviews.append(game_details['reviews'])
    action_game_prices.append(game_details['price'])
    action_game_installs.append(game_details['installs'])
    action_game_developers.append(game_details['developer'])
    action_game_urls.append(game_details['url'])
    i += 1
```

```python
# now we can put those lists into a dataframe
free_action_games = {
    "Game Title": action_game_titles,
    "Game Score": action_game_scores,
    "Number of Reviews": action_game_reviews,
    "Price": action_game_prices,
    "Number of Installs": action_game_installs,
    "Developer": action_game_developers,
    "Play Store URL": action_game_urls
    
}
free_action_games_df = pd.DataFrame(free_action_games)
free_action_games_df.head(20)
```
